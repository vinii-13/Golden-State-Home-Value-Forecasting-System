```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('California_Housing_Prices.csv')
```

```python
df
```

```python
df.info()
```

```python
# data preprocessing
# 1. removing null values

df.dropna(inplace = True)
```

```python
df.info()
```

```python
import sklearn
from sklearn.model_selection import train_test_split

# asigning val to x and y variables

x = df.drop(['median_house_value'], axis =1)
y = df['median_house_value']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state=2)
```

```python
# data set to be used for training purpose

train_data = x_train.join(y)
train_data
```

```python
train_data = train_data.join(pd.get_dummies(train_data.ocean_proximity, dtype = 'int64')).drop(['ocean_proximity'], axis = 1)
train_data
```

```python
# finding correlation between variables

t = train_data.corr().round(2)
```

```python
plt.figure(figsize = (10,7))
sns.heatmap(data = t, annot = True ,cmap = 'YlGnBu')
```

Inference: we find a positive correlation between median_income and median_house_value

```python
# lets visualize this more for better understanding

plt.figure(figsize = (10,5))
sns.scatterplot(x = "latitude", y = "longitude", data = train_data , hue = 'median_house_value', palette= "coolwarm")
```

Inferences: negative correlation between variables

```python
# visualizing correlation between median_house_value and median_income

plt.figure(figsize = (10,15))

plt.scatter(df.median_income, df.median_house_value, marker = 'x')
```

Inferences: npositive correlation between variables

```python
# importing linear regression

from sklearn.linear_model import LinearRegression
lreg = LinearRegression()

x_train, y_train = train_data.drop(['median_house_value'], axis = 1), train_data['median_house_value']
lreg.fit(x_train,y_train)
```

```python
test_data = x_test.join(y_test)
test_data = test_data.join(pd.get_dummies(test_data.ocean_proximity, dtype = 'int64')).drop(['ocean_proximity'], axis = 1)
# test_data = test_data.join(pd.get_dummies(test_data.ocean_proximity, dtype = 'int64')).drop(['ocean_proximity'], axis = 1)
test_data
```

```python
x_test, y_test =  test_data.drop(['median_house_value'], axis = 1), test_data['median_house_value']
```

```python
lreg.score(x_test,y_test)
```

Inference: Accuracy score of this model is not so good, hence we improvise the result by using random forest model

```python
from sklearn.ensemble import RandomForestRegressor

forest = RandomForestRegressor()
x_train, y_train = train_data.drop(['median_house_value'], axis = 1), train_data['median_house_value']
```

```python
forest.fit(x_train,y_train)
```

```python
x_test, y_test =  test_data.drop(['median_house_value'], axis = 1), test_data['median_house_value']
forest.score(x_test,y_test)
```

Our model's acuracy has improved and it now ready to execute.
